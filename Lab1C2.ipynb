```python

import pandas as pd
import matplotlib.pyplot as plt
from numpy import where
import openpyxl
wb = openpyxl.Workbook()
```

```python

archivo = pd.read_csv("/trx_tarjetas_credito.csv")
#print(archivo, archivo["Amount"])

procesar = pd.crosstab(index=archivo["Gender"],columns="cant_veces_repite")
#print(procesar)

agrupando = archivo.groupby(["Amount"]).agg({lambda x:x+","})
#print(agrupando)

plt.pie(procesar, labels=procesar.index)
plt.xlabel("Monto Total")
plt.show()
wb.save('Genero.xlsx')
```
